```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
returns = pd.DataFrame(np.random.normal(1.0, 0.03, (100, 10)))

prices = returns.cumprod()

prices.plot()
plt.title('Randomly-generated Prices')
plt.xlabel('Time')
plt.ylabel('Price')
plt.legend(loc = 0);
```
